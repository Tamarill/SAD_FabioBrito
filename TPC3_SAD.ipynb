<h1>Importação dataset</h1>

```
import sklearn.datasets as datasets
import pandas as pd
import numpy as np
data = datasets.california_housing.fetch_california_housing()
data
```

<h1>Aggregate</h1>

```
alldata = np.column_stack((data.data,data.target))
allnames = np.append(data.feature_names,"Target")
dframe = pd.DataFrame(alldata, columns=allnames)

stat = dframe.groupby("MedInc").agg(["mean"])
stat
```
